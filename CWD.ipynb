```python
import numpy as np
data=np.arange(1021)
```

```python
for i in range(1021):
  x=0
  cwd=0
  a=df1.iloc[:,i]
  for j in range(365):
    if a[j]>=1:
      x=x+1
      cwd=max(x,cwd)
    else :
      x=0
      cwd=max(x,cwd)
  data[i]=cwd
```
